2. Ingresar por teclado  la fecha actual: dia,mes y año tambien ingresar la fecha de nacimiento de una persona y calcular cuantos años tiene en años, meses y días

```python
dia= int(input("Inserte que dia es hoy:"))
mes= int(input("Inserte que mes es hoy en numeros:"))
año= int(input("Inserte que año es hoy:"))
dia1= int(input("Inserte su dia de nacimiento :"))
mes1= int(input("Inserte su mes de nacimiento en numeros :"))
año1= int(input("Inserte su año de nacimiento :"))


print('Usted tiene:',año-año1,'años',((mes-mes1)+12), 'meses', ((dia-dia1)+31), 'dias')
```

3. Elaborar un programa en python que permite ingresar el radio y la altura de un cilindro y calcular el volumen del cilindro sabiendo que la formula es π$r^2$h

```python
radio=int(input("Ingresar el radio del cilindro"))
altura=int(input("Ingresar la altura del cilindro"))
print('Volumen del cilindro es=',3.141592653589793*(radio**2)*altura,'m³')
```

4. Elaborar un programa que permita convertir una distancia medida en metros a centimetros, milímetros y kilómetros. El usuario debe ingresar el valor en metros y el programa debe mostrar las tres equivalencias correspondientes.

```python
metros=int(input("Ingresar el valor que quiere convertir de metros a mlímetros,centímetros y kilómetros:"))
print('El valor de metros a milimetros es:',metros*1000,'mm')
print('El valor de metros a centímetros es:',metros*100,'cm')
print('El valor de metros a kilometros es:',metros/1000,'km')
```
