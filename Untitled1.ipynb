Ingresar dos valores enteros por teclado y muestre por pantalla su división si el divisor es cero el programa debe mostrar un mensaje de error.

```python

x=float(input("Ingrese el dividendo:"))
y=float(input('Ingrese el "divisor:'))
if y==0:
     print('División para cero no existe')
else:
    print('El resultado de la división es:', x/y)
```
