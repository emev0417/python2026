1. Escribe un programa que pida el peso (en kg) y la estatura (en metros) y calcule el IMC con la fórmula: Los datos de peso y altura deben ser ingresados por teclado

$$IMC=\frac{Peso}{estatura^2}$$
	​


```python
peso = float(input("Ingrese su peso:")) #Aplico la funcion de input y le convierto en flotante con float
estatura= float(input("Ingrese su estatura:"))
imc=peso/pow(estatura,2) #aqui realizo la operacion y aplico pow para la potencia
print(imc) #imprimo el resultado
```

Escribe un programa que pida las coordenadas de dos puntos en el plano:
$$(x_1,y_1) y (x_2 , y_2)$$

Y calcule la distacia entre ellos
$$distancia=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$

Los datos x1,x2,y1,y2 son ingresados por teclado. Pedir al profesor ayuda para utilizar la función sqrt de la raiz


```python
x1=2
x2=3
y1=1
y2=4
x=x2-x1 #Realizo la resta x
y=y2-y1 #Realizo la resta y
operacion=(pow(x,2))+(pow(y,2)) #Redlizo la operacion dentro de la raíz
import math #importo la biblioteca
raiz=math.sqrt(operacion) #realizo la operacion de raiz con sqrt
print('La distancia es',(raiz)) #imprimo la respuesta
```
