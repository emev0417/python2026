### 1.5 Paso de parámetros
* Tipos simples (inmutables) por valor: int, float, string.
    *  El efecto es como si se creara una copia dentro de la función, aunque realmente no es esto lo que ocurre.
    *   Los cambios dentro de la función no afectan fuera.

```python
def cambiar_valor(x):
    x += 3
    print('Dentro de la función:', x) # x ahora referencia a un nuevo objeto (el 3), pero esto no afecta a la variable de fuera de la función, que sigue apuntando al 2.
    return
   
a = 2
cambiar_valor(a) #Aqui utiliza la funcion 
print('Fuera de la función:', a) #Aqui no por eso sale el valor de a=2
```

### 1.8 Funciones Lambda
* Es posible definir funciones anónimas (sin nombre).
* Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def
no puede (por ejemplo, dentro de una lista o como parámetro de una función).
* Útiles cuando se quiere pasar una función como argumento a otra.<br>
Formato general:<br>
`lambda <lista de argumentos> : <valor a retornar>`
* Importante: <valor a retornar> no es un conjunto de instrucciones. Es simplemente una
expresión return que omite esta palabra.
* Las funciones definidas con def son más generales:
    * Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con def, pero no viceversa.

```python
# Ejemplo: función suma.
def suma(x, y, z):
    return x + y + z
print(suma(1, 2, 3))

# Ejemplo: función suma como expresión lambda.
suma2 = lambda x, y, z : x + y + z
print(suma2(1, 2, 3))
```

```python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
cities.sort()
print(cities)
cities.sort(key = lambda x : len(x))
print(cities)
cities.sort(key = len)
print(cities)
def longitud(x):
    return len(x)
cities.sort(key = longitud)
print(cities)
  
```

```python
# Mismo ejemplo sin lambda
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
def num_caracteres(x):
    return len(x)
cities.sort(key = num_caracteres)
print(cities)
```

* Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
* Ideal para escoger entre varias opciones

```python
def switch_dict(operator, x, y):
# crea un diccionario de opciones y selecciona 'operator'
    return {
        'add': lambda: x + y,
        'sub': lambda: x - y,
        'mul': lambda: x * y,
        'div': lambda: x / y,
    }.get(operator, lambda: None)() # retorna None si no encuentra 'operator'
print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
```

### 1.9 Scopes
* La localización de una asignación a una variable en el código determina desde donde puedes
acceder a esa variable.
* Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
* El alcance de la visibilidad de una variable determina su scope.
* El término scope hace referencia a un espacio de nombres (namespace). Por ejemplo, una función establece su propio namespace.

```python
X = 10
def func():
        X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'
        def func_int():
            X = 30
            print(X)
        func_int()
func()
```

**Regla LEGB** Dada una función F, tenemos los siguientes scopes:
* **L**ocal: variables locales a F.
* **E**nclosing: variables localizadas en funciones que contienen a (o por encima de) F.
* **G**lobal: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este scope no abarca más de un fichero.
* **B**uilt-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

```python
#Global (X y func)
X = 99
def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z
print(func(1))
```

```python
X = 1
def func():
    X = 2 # X es una variable diferente
func()
print(X)
```

* La sentencia global nos permite modificar una variable global desde dentro de una función.


```python
X = 1
def func():
    global X
    X = 2
func()
print(X)
```

* No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.

```python
y, z = 1, 2
def todas_globales():
    global x
    x = y + z # Las tres variables son globales.
todas_globales()
print(x)
print(y)
print(z)
```

```python
X = 77
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    f2()
f1()
```

**Closures**
* Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan
existiendo o no.

```python
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2 # Devuelve f2, sin invocarla
    
accion = f1()
accion()
```

* Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales.

```python
def f1():
    contador = 10
    def f2():
        nonlocal contador
        contador += 1
        print(contador)
    return f2
accion = f1()
accion()
accion()
accion()
```

1. Escribe una función que reciba como entrada una lista con números y devuelva como resultado una lista con los cuadrados de los números contenidos en la lista de entrada.

```python
def calcular_cuadrados(lista_numeros):
    # 1. Creamos una lista vacía para guardar los resultados
    lista_resultados = []
    
    # 2. Recorremos la lista de entrada con un bucle for
    for n in lista_numeros:
        # Calculamos el cuadrado (n elevado a la 2)
        cuadrado = n ** 2
        # Agregamos el resultado a nuestra lista
        lista_resultados.append(cuadrado)
        
    # 3. Devolvemos la lista final
    return lista_resultados

# --- Prueba del programa ---
mi_lista = [2, 4, 6, 8]
resultado = calcular_cuadrados(mi_lista)
print("Los cuadrados son:", resultado)
```

2. Escribe una función que reciba números como entrada y devuelva la suma de los mismos. La
función debe ser capaz de recibir una cantidad indeterminada de números. La función no
debe recibir directamente ningún objeto complejo (lista, conjunto, etc.).

```python
def sumar_todos(*numeros):
    # 'numeros' se comporta como una tupla que contiene todos los valores
    total = 0
    
    # Usamos un bucle for para recorrer cada número recibido
    for n in numeros:
        total = total + n
        
    return total

# --- Ejemplos de uso ---
# Podemos pasarle 3 números...
resultado1 = sumar_todos(10, 20, 30)
print("Suma de tres números:", resultado1)

# ...o podemos pasarle 6 números o los que queramos
resultado2 = sumar_todos(1, 2, 3, 4, 5, 6)
print("Suma de seis números:", resultado2)
```

2.  No recibe objetos complejos: Fíjate que al llamar a la función no escribimos corchetes [] ni llaves {}. Pasamos los números "sueltos", separados por comas.
3.  El bucle for: Es la forma más limpia de ir sumando cada elemento uno por uno en la variable total.

* Un pequeño consejo de examen:
Si tu profesor te pregunta qué tipo de dato es numeros dentro de la función, la respuesta técnica es una Tupla. Las tuplas son muy parecidas a las listas, pero no se pueden modificar, lo cual es perfecto para solo leer y sumar los valores.

```python
#Este ejercicio es otro no de los del profe
def hallar_maximo(*numeros):
    # Si no nos pasan números, devolvemos un mensaje o None
    if len(numeros) == 0:
        return "No se proporcionaron números"

    # 1. Suponemos que el primer número es el más grande
    mayor = numeros[0]
    
    # 2. Comparamos con el resto de los números
    for n in numeros:
        if n > mayor:
            # Si encontramos uno más grande, actualizamos la variable
            mayor = n
            
    return mayor

# --- Pruebas ---
print("El mayor es:", hallar_maximo(5, 12, 8, 45, 3))
print("El mayor es:", hallar_maximo(-10, -2, -5))
```

3. Escribe una función que reciba un string como entrada y devuelva el string al revés. Ejemplo:
si el string de entrada es ‘hola’, el resultado será ‘aloh’.

```python
def invertir_cadena(texto):
    # 1. Creamos un string vacío para ir guardando el resultado
    resultado = ""
    
    # 2. Recorremos el texto original letra por letra
    for letra in texto:
        # Colocamos la letra al PRINCIPIO de lo que ya llevamos
        # Esto hace que la palabra se vaya armando al revés
        resultado = letra + resultado
        
    return resultado

# --- Prueba ---
palabra = "hola"
print(f"El original es: {palabra}")
print(f"Al revés es: {invertir_cadena(palabra)}")
```

6. Escribe una función que reciba un número entero positivo como parámetro y devuelva una
lista que contenga los 5 primeros múltiplos de dicho número. Por ejemplo, si la función recibe
el número 3, devolverá la lista [3, 6, 9, 12, 15]. Si la función recibe un parámetro incorrecto
(por ejemplo, un múmero menor o igual a cero), mostrará un mensaje de error por pantalla
y devolverá una lista vacía.


```python
def obtener_cinco_multiplos(numero):
    # 1. Validación: Comprobamos si el número es positivo
    if numero <= 0:
        print("Error: El número debe ser un entero positivo (mayor que cero).")
        return [] # Devolvemos lista vacía como pide el enunciado

    # 2. Si es válido, creamos la lista de múltiplos
    multiplos = []
    
    # Usamos un bucle que se repita 5 veces (del 1 al 5)
    for i in range(1, 6):
        resultado = numero * i
        multiplos.append(resultado)
        
    return multiplos

# --- Pruebas del programa ---
print("Múltiplos de 3:", obtener_cinco_multiplos(3))
print("Múltiplos de 10:", obtener_cinco_multiplos(10))
print("Prueba con error (0):", obtener_cinco_multiplos(0))
```

* ¿Cómo explicar este código ante tu clase?
El Filtro Inicial: Lo primero que hace la función es una "guarda". Si el número no nos sirve (<= 0), mostramos el error y usamos return []. El return corta la función ahí mismo, por lo que el resto del código no se ejecuta si el parámetro es incorrecto.

* El Bucle range(1, 6): Genera los números 1, 2, 3, 4, 5. Es la forma más fácil de obtener los multiplicadores que necesitamos.

* La Acumulación: En cada vuelta del bucle, multiplicamos el número original por el contador i y "guardamos" ese resultado en la lista usando .append().

```python
def obtener_cinco_multiplos(numero):
    if numero <= 0:
        print("Error: Parámetro incorrecto.")
        return []
    
    # Creamos la lista en una sola línea
    return [numero * i for i in range(1, 6)]
```

¿Te gustaría que hiciéramos un ejercicio donde el usuario pueda decidir cuántos múltiplos quiere ver en lugar de que sean siempre 5? Sería añadir un segundo parámetro a la función.

```python
def obtener_multiplos_personalizados(numero, cantidad):
    # 1. Validamos que ambos números sean positivos
    if numero <= 0 or cantidad <= 0:
        print("Error: Ambos valores deben ser enteros positivos.")
        return []

    # 2. Generamos la lista usando una List Comprehension
    # El rango irá desde 1 hasta cantidad + 1
    multiplos = [numero * i for i in range(1, cantidad + 1)]
    
    return multiplos

# --- Pruebas ---
print("Los primeros 3 múltiplos de 7:", obtener_multiplos_personalizados(7, 3))
print("Los primeros 10 múltiplos de 2:", obtener_multiplos_personalizados(2, 10))
```

7. Escribe una función que reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá True si la lista tiene duplicados y False si no los tiene.


```python
def tiene_duplicados(lista):
    # 1. Convertimos la lista en un conjunto (set)
    # Al hacer esto, Python elimina automáticamente los repetidos
    conjunto_sin_repetidos = set(lista)
    
    # 2. Comparamos los tamaños (longitudes)
    # Si el tamaño cambió, significa que había duplicados
    if len(conjunto_sin_repetidos) < len(lista):
        return True
    else:
        return False

# --- Pruebas ---
lista_a = [1, 2, 3, 4, 5]
lista_b = [1, 2, 3, 2, 5]

print(f"¿La lista A tiene duplicados?: {tiene_duplicados(lista_a)}") # False
print(f"¿La lista B tiene duplicados?: {tiene_duplicados(lista_b)}") # True
```

```python
def tiene_duplicados_bucle(lista):
    vistos = []
    for elemento in lista:
        if elemento in vistos:
            return True
        vistos.append(elemento)
    return False
# --- Pruebas ---
lista_a = [1, 2, 3, 4, 5]
lista_b = [1, 2, 3, 2, 5]

print(f"¿La lista A tiene duplicados?: {tiene_duplicados(lista_a)}") # False
print(f"¿La lista B tiene duplicados?: {tiene_duplicados(lista_b)}") # True
```

8. Escribe una función lambda que, al igual que la función desarrollada en el ejercicio anterior,
reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá
True si la lista tiene duplicados y False si no los tiene.

```python
# Definimos la lambda
# 'lista' es el parámetro de entrada
# La expresión compara si el tamaño del set es menor al de la lista original
comprobar_duplicados = lambda lista: len(set(lista)) < len(lista)

# --- Pruebas del programa ---
lista_1 = [1, 2, 3, 4]
lista_2 = [1, 2, 2, 3]

print(f"¿Lista 1 tiene duplicados?: {comprobar_duplicados(lista_1)}") # False
print(f"¿Lista 2 tiene duplicados?: {comprobar_duplicados(lista_2)}") # True
```

9. Escribe una función que compruebe si un string dado es un palíndromo. Un palíndromo es
una secuencia de caracteres que se lee igual de izquierda a derecha que de derecha a izquierda.
Por ejemplo, la función devolverá True si recibe el string “reconocer” y False si recibe el string
“python”.


```python
def es_palindromo(texto):
    # 1. Limpiamos el texto: lo pasamos a minúsculas
    # Esto evita errores si el usuario escribe "Reconocer" (con R mayúscula)
    texto = texto.lower()
    
    # 2. Invertimos el texto usando el truco del slicing [::-1]
    texto_invertido = texto[::-1]
    
    # 3. Comparamos y devolvemos el resultado
    if texto == texto_invertido:
        return True
    else:
        return False

# --- Pruebas ---
palabra1 = "reconocer"
palabra2 = "python"

print(f"¿Es '{palabra1}' un palíndromo?: {es_palindromo(palabra1)}") # True
print(f"¿Es '{palabra2}' un palíndromo?: {es_palindromo(palabra2)}") # False
```

¡Excelente! Para que una frase como "Anita lava la tina" sea reconocida como palíndromo, necesitamos un paso de limpieza. Si no quitamos los espacios, Python comparará los espacios en blanco y dirá que no son iguales.

```python
def es_palindromo_frase(frase):
    # 1. Quitamos los espacios y pasamos a minúsculas
    # El método .replace(" ", "") busca espacios y los elimina
    limpio = frase.replace(" ", "").lower()
    
    # 2. Comparamos con su versión invertida
    return limpio == limpio[::-1]

# --- Pruebas ---
frase1 = "Anita lava la tina"
frase2 = "Sometamos o matemos"

print(f"¿'{frase1}' es palíndromo?: {es_palindromo_frase(frase1)}") # True
print(f"¿'{frase2}' es palíndromo?: {es_palindromo_frase(frase2)}") # True
```
