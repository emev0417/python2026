```python
# --- EJERCICIO DE LITERALES COMPUESTOS ---

# 1. LISTA (Mochila): La podemos cambiar
mi_lista = ["Manzana", "Pera"]
mi_lista[0] = "Piña"  # Cambiamos Manzana por Piña
print("Lista modificada:", mi_lista)

# 2. TUPLA (Caja sellada): ¡Dará error si intentas cambiarla!
mi_tuple = ("Manzana", "Pera")
print("Tupla original:", mi_tuple)
# Si quitas el '#' de la siguiente línea, verás el error:
# mi_tuple[0] = "Piña" 

# 3. DICCIONARIO (Agenda): Buscamos por nombre
mi_dict = {"nombre": "Eme", "materia": "Python"}
print("Dato específico del diccionario:", mi_dict["nombre"])

# 4. CONJUNTO (Saco): Quita el duplicado automáticamente
mi_set = {1, 2, 2, 3, 3, 3}
print("Conjunto sin repetidos:", mi_set)
```
