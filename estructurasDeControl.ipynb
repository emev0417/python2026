## 1 Estructuras de control
1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’

```python
# Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.

```python
x = -3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

```python
#Utilizo float ya que la division puede o no ser exacta y la respuesta seria con decimales.
x=float(input("Ingrese el dividendo:"))
y=float(input('Ingrese el "divisor:'))

#Uso if para que si "y" es igual 0 de error 
if y==0:
     print('División para cero no existe')
    
#De lo contrario realiza normalmente la división
else:
    print('El resultado de la división es:', x/y)
```

2. Escribir un programa en python que pida al usuario un número entero por teclado y muestre en pantalla si es par o impar.

```python
#Con la función int hacemos que lo que ingrese a input cambie de tipo de string a entero.
numero=int(input("Ingrese un número entero: "))

#Aqui el if verifica si el numero ingresado tiene residuo es impar sino es par.
if numero % 2==0: 
    print("El número es par")
else:
    print("El número es impar")
```

Escribir un programa en python para una empresa que tiene salas de juego para todas las edades y quiere calcular de forma automática el precio que debe cobrar a sus clientes por entrada.El programa debe preguntar al usuario: el nombre y la edad del cliente y mostrar el precio de entrada si el cliennte es menor de 4 años puede ingresar gratis, si tiene entre 4 y 18 años debe pagar 5 dólares y si es mayor de 18 años 10  dólares.

```python
name = input("Ingrese su nombre:")

#Convertimos la funcion input en entero con int
age = (int(input("Ingrese su edad:")))

#Aqui el if verifica si el numero ingresado es menor a 4
if age<4:
    print('Su entrada es gratuita')

#Aqui el elif verifica si el numero ingresado es mayor igual 4 y menor que 18
elif age>=4 and age<=18:
    print('El valor de su entrada es 5$')

#De lo contrario el numero es mayor de 18
else:
    print('El valor de su entrada es 10')
       
```
