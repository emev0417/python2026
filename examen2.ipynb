4. Escribir un programa en python donde una función reciba una frase y cuente cuantas vocales tiene la frase.

```python
#Aqui usamos def para definir la funcion
def contar_vocales(frase):
#Aqui empieza en ceropara que vaya sumando cada que encuentra una vocal
    contador= 0
#Usamos for para que vaya viendo frase
    for letra in frase:
#Con if revisa si hay vocales
        if letra in "aeiou":
#Si encuentra una vocal la suma al contador
            contador += 1
    return contador
#Aqui usamos la función
texto = input("ingrese una frase:")
print("numero de vocales:", contar_vocales(texto))
        
```
